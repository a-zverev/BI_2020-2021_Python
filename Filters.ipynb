```python
from PIL import Image
import numpy as np
from io import BytesIO
import requests
from random import randint
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(rc={'figure.figsize': (11, 8)})
%matplotlib inline
```

```python
url = 'https://avatars.githubusercontent.com/a-zverev'
img = Image.open(requests.get(url, stream=True).raw)
data = np.array(img)

img
```

```python
print(data.shape)
```

```python
def convolve(img, kernel):
    """
    "Convolution" of img with kernel, assuming stride is equal
    to 1
    """
    # Assuming kernel is 3 x 3 and stride equals to 1
    modified = np.zeros_like(img)
    for row in range(1, img.shape[0] - 2):
        for col in range(1, img.shape[1] - 2):
            modified[row, col] = (img[row:row + 3, col:col + 3]
                                  * kernel).mean()
    return modified


def generate_kernel():
    res = np.random.randint(3, size=(3, 3))
    res = res - 1
    return res
```

```python
generate_kernel()
```

```python
kernel = generate_kernel()
plt.imshow(convolve(data, kernel), cmap='gray')
plt.axis('off')
plt.show()
print(kernel)
```
